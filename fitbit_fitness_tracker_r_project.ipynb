# 1. Loading packages

```r
library(tidyverse)
library(lubridate)
library(dplyr)
library(ggplot2)
library(tidyr)
```

# 2. Importing datasets
Disini menggunakan FitBit Fitness Tracker [Data](https://www.kaggle.com/datasets/arashnic/fitbit).

```r
# Import files:

activity <- read.csv("dailyActivity_merged.csv")
sleep <- read.csv("sleepDay_merged.csv")
weight <- read.csv("weightLogInfo_merged.csv")
```

# 3. Explore datasets

```r
# Amati baris pertama dari data:

head(activity)
head(sleep)
head(weight)
```

```r
# Cek nama kolom di file

colnames(activity)
colnames(sleep)
colnames(weight)
```

```r
# Berapa banyak unique partisipasi pada dataframe ?

n_distinct(activity$Id)
n_distinct(sleep$Id)
n_distinct(weight$Id)
```

Conclusion:
1. Dataset "activity" memiliki 33 partisipan, dataset "sleep" memiliki 24 partisipan dan "weight" hanya memiliki 8 partisipan. Disini tidak dapat menggunakan dataset "weight". Jumlah partisipan yang dimilikinya tidak cukup untuk menarik kesimpulan.

# 4. Clean datasets and merge

```r
# Menghapus kolom yang tidak perlu

activity <- activity %>%
  select(-c("TrackerDistance" , "LoggedActivitiesDistance", "SedentaryActiveDistance"))

sleep <- sleep %>%
  select(-c("TotalSleepRecords"))
```

```r
# Mengganti nama kolom dalam set data (untuk membuatnya konsisten sebelum penggabungan):

activity <- rename(activity, Date = ActivityDate)
sleep <- rename(sleep, Date = SleepDay)
```

```r
# Cek tipe data di dataset

str(activity)
str(sleep)
```

```r
# Mengubah tipe data di kolom "Date" pada dataset

activity$Date=as.Date(activity$Date, format="%m/%d/%y")
sleep$Date=as.Date(sleep$Date, format="%m/%d/%y")
```

```r
# Periksa jumlah observasi dalam set data:

nrow(activity)
nrow(sleep)
```

```r
# Cek baris data duplicated di dataset

nrow(activity[duplicated(activity), ])
nrow(sleep[duplicated(sleep), ])
```

```r
# Menghapus baris duplicate di dataset 'sleep':

sleep <- sleep %>% distinct()
```

```r
# Cek hasil:

nrow(sleep)
```

```r
# Merge datasets:

merged_data <- merge(sleep, activity, by=c('Id', 'Date'))
```

```r
# Observasi dari baris pertama pada data merged

head(merged_data)
```

# 5. Analyze the data

```r
# Observasi beberapa tren umum dalam data:

hist(merged_data$TotalSteps)
hist(merged_data$VeryActiveMinutes)
hist(merged_data$SedentaryMinutes)
hist(merged_data$Calories)
```

```r
# Cek ringkasan statistik:

merged_data %>%
  select(TotalSteps, TotalDistance, VeryActiveDistance, ModeratelyActiveDistance,
         LightActiveDistance) %>%
  summary()

merged_data %>%
  select(VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes,
         SedentaryMinutes, Calories) %>%
  summary()

merged_data %>%
  select(TotalMinutesAsleep, TotalTimeInBed) %>%
  summary()
```

Observations:
1. Rata-rata sedentary minutes adalah 712,1 menit atau 12 jam.
2. Rata-rata very active minutes adalah 25,05 menit. Rata-rata fairly active minutes adalah 17,92 menit.
3. Rata-rata lighty active minutes adalah 216,5 menit atau 3,6 jam.
4. Rata-rata jumlah kalori yang dibakar per orang per hari adalah sekitar 2.389 kilokalori.


Conclusions:
1. Para peserta sebagian besar tidak aktif sepanjang hari.
2. Para peserta menghabiskan sedikit waktu untuk berolahraga.
3. Para peserta kurang mungkin untuk mengambil bagian dalam kegiatan yang intens/keras.

# 6. Visualization

```r
ggplot(data=activity, aes(x=TotalSteps, y=Calories)) +
  geom_point() + geom_smooth() + labs(title="Total Steps vs. Calories")
```

Conclusion:
1. Terdapat korelasi positif antara Total Langkah dan Kalori, yang mana hal tersebut sudah jelas - semakin aktif, semakin banyak kalori yang di bakar.

```r
ggplot(data=sleep, aes(x=TotalMinutesAsleep, y=TotalTimeInBed)) +
  geom_point()+ labs(title="Total Minutes Asleep vs. Total Time in Bed")
```

Conclusion:
1. Hubungan antara Total Menit Tidur dan Total Waktu di Tempat Tidur terlihat linear. Jadi, jika pengguna Bellabeat ingin meningkatkan kualitas tidur mereka, mereka sebaiknya mempertimbangkan untuk menggunakan notifikasi untuk pergi tidur.

```r
ggplot(data=merged_data, aes(x=TotalMinutesAsleep, y=SedentaryMinutes)) +
geom_point(color='darkblue') + geom_smooth() +
  labs(title="Minutes Asleep vs. Sedentary Minutes")
```

Conclusion:
1. Di sini kita dapat dengan jelas melihat hubungan negatif antara Sedentary Minutes dan Waktu Tidur.

2. Sebagai ide: jika pengguna Bellabeat ingin meningkatkan kualitas tidur mereka, aplikasi Bellabeat dapat merekomendasikan untuk mengurangi reducing sedentary time.

3. Perlu diingat bahwa kita perlu mendukung wawasan ini dengan lebih banyak data, karena korelasi antara beberapa data tidak berarti sebab-akibat.
